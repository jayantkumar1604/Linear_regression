```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
```

```python
new_df= df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,df['per capita income (US$)'])
```

```python
model.predict([[2020]])
```

```python
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+', label='Training data')
plt.plot(df.year, model.predict(new_df), color='blue', label='Regression line')
plt.scatter(2020, model.predict([[2020]]), color='green', s=100, label='Prediction 2020')
plt.text(2020, model.predict([[2020]])[0], f"${model.predict([[2020]])[0]:.2f}", color='green')

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
